```python
nan_columns = five_second_data.columns[five_second_data.isnull().any()].to_list()
len(nan_columns)
```

```python
not_null_col = [col for col in five_second_data.columns if five_second_data[col].isnull().sum()<800]
len(not_null_col)
```

```python
new_data = pd.DataFrame(five_second_data[not_null_col])
new_data.head()
```

```python
not_null = [col for col in new_data.columns if new_data[col].isnull().sum() < 1]
len(not_null)
```

```python
not_null_data = new_data[not_null]
```

```python
new__null = [i for i in new_data.columns if 1<= new_data[i].isnull().sum()<800]
len(new__null)
```

```python
null_data = new_data[new__null].rolling(window=10, min_periods=1).mean()
```

```python
final_data = pd.concat([not_null_data, null_data], axis=1)
```

```python
final_data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
target_num = final_data['target'].unique()
target_num
```

```python
user_num = final_data['user'].unique()
```

```python
oe = OrdinalEncoder(categories=[target_num, user_num])
oe.fit(final_data[['target','user']])
```

```python
new_target = oe.transform(final_data[['target','user']])
new_target_data = pd.DataFrame(new_target, columns=['target', 'user'])
```

```python
final_data.head()
```

```python
x = final_data.drop(['target', 'user', 'Unnamed: 0'], axis=1)
y = new_target_data.iloc[:,0]
```

```python
data = pd.concat([x,y], axis=1)
```

```python
from sklearn.tree import ExtraTreeClassifier
extra_tree = ExtraTreeClassifier()
```

```python
extra_tree.fit(x,y)
```

```python
extra_tree.feature_importances_
```

```python
importance_feat = pd.Series(extra_tree.feature_importances_, index=x.columns)
plt.figure(figsize=(16,12))
importance_feat.nlargest(40).plot(kind = 'barh')
plt.show()
```
