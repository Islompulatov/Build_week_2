```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
```

```python
df = pd.read_csv('test_data.csv', index_col=None)
df.head()
```

```python
model = joblib.load('model_joblib')
```

```python
y = df['target']
x = df.drop(['user', 'target', 'Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```

```python
x.head()
```

```python
model.fit(x, y)
pred = model.predict(x)
acc = accuracy_score(y, pred)
acc
```
